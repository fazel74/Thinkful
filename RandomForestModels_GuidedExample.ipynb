## PCA:

```python
# Import and run PCA model 
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
# Test PCA on forest
rfc = ensemble.RandomForestClassifier()
cross_val_score(rfc, X_pca, Y, cv=10)
```

```python
pca = PCA(n_components=12)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
import warnings
warnings.filterwarnings('ignore')
# Test PCA on forest
rfc = ensemble.RandomForestClassifier()
cross_val_score(rfc, X_pca, Y, cv=10)
```

Using PCA dimensionality reduction, almost 90% accuracy can be maintained by reducing the data to 20 principle components. Less than 20 will result for sure in <90% accuracy.
