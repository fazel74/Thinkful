# Quick Sort

```python
import time
import random

# Set seed.
random.seed(a=100)

# Create our default list.
short_list = list(random.sample(range(1000000), 10))
long_list = list(random.sample(range(1000000), 10000))
```

```python
def qsort_intern(array, start, end):
    if start < end: # stop recrusion - for example for [1], we dont need to sort it 
                    
        pivot = end 
        th = start  
        for i in range(start,end): # function compare all elements with pivot value
            if array[i] < array[pivot]:
                array[th],array[i] = array[i], array[th] # swap elements 
                th += 1 # move treshold
        if array[pivot] < array[th]: 
            array[pivot], array[th] = array[th], array[pivot] 
        
        qsort_intern(array, start, th - 1) # for array under treshold
        qsort_intern(array, th + 1, end)   # for array above treshold
    
def QuickSort(array):
    start = 0
    end = len(array) - 1
    return qsort_intern(array, start, end)
```

```python
array = [5,3,7,2,3,7,5,8,8,4,1,2,3,5]
# Start the timer.
start_time = time.time()

QuickSort(array)

print("--- %s seconds ---" % (time.time() - start_time))
print(array)
```

```python
# Start the timer.
start_time = time.time()

QuickSort(short_list)

print("--- %s seconds ---" % (time.time() - start_time))
print(short_list)
```

```python
# Start the timer.
start_time = time.time()

QuickSort(long_list)

print("--- %s seconds ---" % (time.time() - start_time))
print(long_list)
```
