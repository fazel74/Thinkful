### Observing the factors that lead to employee attrition based on collected data in IBM Human Resource Department. 

The Data-Set consist of IBM Employee's informoation to study why workers quit their jobs. Attrition tells the status of employee and where attrition is yes means those who leave organisation what is the state of them means their Monty income, their life Balance before they leave the organisation, and etc..

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Reading data and taking a look at it and dataframe structure:

```python
df = pd.read_csv('HR-Employee-Attrition.csv')

df.head()
```

## Question 1: Can distance from home by work place influence on the employee attrition? 

```python
# changing yes and no values to 1 and 0
df['Attrition'].loc[df['Attrition'] == 'Yes'] = 1
df['Attrition'].loc[df['Attrition'] == 'No'] = 0

df_d_a_y = df['DistanceFromHome'].loc[df['Attrition'] == 1]
df_d_a_n = df['DistanceFromHome'].loc[df['Attrition'] == 0]

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.hist(df['DistanceFromHome'])
plt.ylabel('Numbers')
plt.xlabel('Distance from home by work place')
plt.title('Distribution of all employees')

plt.subplot(1, 2, 2)
plt.hist(df_d_a_y, color='purple')
plt.ylabel('Numbers')
plt.xlabel('Distance from home by work place')
plt.title('Distribution of employees with Attrition')


plt.tight_layout()
plt.show()
```

Comparing data distribution where there is Attrition with all data, we can see a relatively increase in the area of distances more than 15 miles. 

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.boxplot(df_d_a_y)
plt.ylabel('DistanceFromHome')
plt.title('Attrition = Yes')

plt.subplot(1, 2, 2)
plt.boxplot(df_d_a_n)
plt.ylabel('DistanceFromHome')
plt.title('Attrition = No')

plt.tight_layout()
plt.show()
```

Observing boxplots where Attrition = Yes and whre Attrition = No, we can conclude that the distance from hom by work place associates with employee attrition. Employee are more likely to quit , When DistanceFromHome is generally above 10 miles.

# Question 2: Is there an association between employee attrition and his or her level of education?

```python
df_e_a_y = df['Education'].loc[df['Attrition'] == 1]
df_e_a_n = df['Education'].loc[df['Attrition'] == 0]

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.hist(df['Education'], color='red', alpha=0.5)
plt.hist(df_e_a_n, color='blue', alpha=0.5)
plt.ylabel('Numbers')
plt.xlabel('Education 1:Below College 2:College 3:Bachelor 4:Master 5:Doctor')
plt.title('Attrition = No')

plt.subplot(1, 2, 2)
plt.hist(df['Education'], color='purple', alpha=0.5)
plt.hist(df_e_a_y, color='green', alpha=0.5)
plt.ylabel('Numbers')
plt.xlabel('Education 1:Below College 2:College 3:Bachelor 4:Master 5:Doctor')
plt.title('Attrition = Yes')


plt.tight_layout()
plt.show()
```

Employee's level of education is an important factor associating with attrition. It can be seen that employees with lower levels of education relatively leave the company more than those with higer levels of education.

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(df_e_a_y)
plt.ylabel('Education')
plt.title('Attrition = Yes')

plt.subplot(1, 2, 2)
plt.plot(df_e_a_n)
plt.ylabel('Education')
plt.title('Attrition = No')

plt.tight_layout()
plt.show()
```

Although the most data concentrated in center of level of education, observing relatively it can be concluded that lower levels of education associates more than the higher levels with attrition.

# Question 3: Is the factor of 'Years at Company' an important factor impacting employee attrition?

```python
df_y_a_y = df['YearsAtCompany'].loc[df['Attrition'] == 1]
df_y_a_n = df['YearsAtCompany'].loc[df['Attrition'] == 0]


plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.scatter(x=df['YearsAtCompany'], y=df['Attrition'])
plt.ylabel('0: Attrition = No, 1: Attrition = Yes')
plt.xlabel('Years at Company')
plt.title('Attrition based on Years at Company')

plt.subplot(1, 2, 2)
plt.hist(df['YearsAtCompany'], color='red', alpha=0.5)
plt.hist(df_y_a_y, color='blue', alpha=0.5)
plt.ylabel('Numbers')
plt.xlabel('Years at Company')
plt.title('Attrition based on Years at Company')


plt.tight_layout()
```

Certainly 'Years at Company' is a major factor affecting employee attrition. In fact, employees with less years at company often leave the company, and employees with more years at company stay in company.  

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.boxplot(df_y_a_y)
plt.ylabel('Years at Company')
plt.title('Attrition = Yes')

plt.subplot(1, 2, 2)
plt.boxplot(df_y_a_n)
plt.ylabel('Years at Company')
plt.title('Attrition = No')

plt.tight_layout()
plt.show()
```

Observing these two boxplots we can see that employee attrition mostly happens in early years in the company. 

## Conclusion:
Observations from above graphical representations: Attrition rate is higher in distance from home by workplace for distances above 10 miles. It is a factor for job satisfaction, and employees with a high distances between home and work are likely to leave company for finding closer jobs. Another factor that influence on the employee attrition is employee's level of education. Attrition rate is higher for employees with lower levels of education, and amployees with higher levels of education mostly stay and do not quit their jobs. Last factor that is examined in this study is the factor of employee's working years at company. In fact this factor is an important conribution in rate of the employee attrition. Employess who are more years in the company are relatively willing to stay in their workplace, and there is a higher rate of the employee attrition for employees who are in their early years in the company.
