### Observing the factors that lead to employee attrition based on collected data in IBM Human Resource Department. 

The Data-Set consist of IBM Employee's informoation to study why workers quit their jobs. Attrition tells the status of employee and where attrition is yes means those who leave organisation what is the state of them means their Monty income, their life Balance before they leave the organisation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('HR-Employee-Attrition.csv')

df.head()
```

## Question 1: How does distance from home by work place effect on the employee attrition? 

```python
# changing yes and no values to 1 and 0
df['Attrition'].loc[df['Attrition'] == 'Yes'] = 1
df['Attrition'].loc[df['Attrition'] == 'No'] = 0

df2 = df['DistanceFromHome'].loc[df['Attrition'] == 1]

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.hist(df['DistanceFromHome'])
plt.ylabel('Numbers')
plt.xlabel('Distance from home by work place')
plt.title('Distribution of all employees')

plt.subplot(1, 2, 2)
plt.hist(df2, color='purple')
plt.ylabel('Numbers')
plt.xlabel('Distance from home by work place')
plt.title('Distribution of employees with Attrition')



plt.show()
```

```python
plt.scatter(x=df['DistanceFromHome'], y=df['Attrition'], color='green')
plt.ylabel('Attrition')
plt.title('Employee\'s Attrition')
plt.xlabel('Distance from home by work place')
plt.show()
```
