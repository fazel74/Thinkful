```python
import pandas as pd
import re

df = pd.read_csv('apc.csv',encoding = "ISO-8859-1")

df.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)':'Cost (£)'}, inplace=True)
df.columns =[column.replace(" ", "_") for column in df.columns]
df['Cost_(£)'] = df['Cost_(£)'].str.replace('£', '')
df['Cost_(£)'] = df['Cost_(£)'].str.replace('$', '')
df['Cost_(£)'] = df['Cost_(£)'].str.strip()
df['Cost_(£)'] = df['Cost_(£)'].astype(float)
df['Journal_title'] = df['Journal_title'].str.strip()
df['Journal_title'] = df['Journal_title'].str.replace('JOURNAL OF BIOLOGICAL CHEMISTRY', 'Journal of Biological Chemistry')
df['Journal_title'] = df['Journal_title'].str.replace('The journal of Biological Chemistry', 'Journal of Biological Chemistry')
df['Journal_title'] = df['Journal_title'].str.replace('The Journal of Biological Chemistry', 'Journal of Biological Chemistry')
df['Journal_title'] = df['Journal_title'].str.replace('Neuroimage: Clinical', 'Neuroimage')
df['Journal_title'] = df['Journal_title'].str.replace('PNAS', 'Proceedings of the National Academy of Sciences')
#df['Journal_title'] = df['Journal_title'].str.replace('PNAS (Proceedings of the National Academy of Sciences of the United States of America', 'Proceedings of the National Academy of Sciences')
df['Journal_title'] = df['Journal_title'].str.replace('Proceddings of the National Academy of Sciences of USA', 'Proceedings of the National Academy of Sciences')
df['Journal_title'] = df['Journal_title'].str.replace('PNAS USA', 'Proceedings of the National Academy of Sciences')
df['Journal_title'] = df['Journal_title'].str.replace('PLoS 1', 'PLoS ONE')
df['Journal_title'] = df['Journal_title'].str.replace('PLOS  ONE', 'PLoS ONE')
df['Journal_title'] = df['Journal_title'].str.replace('PloS One', 'PLoS ONE')
df['Journal_title'] = df['Journal_title'].str.replace('PLOS One', 'PLoS ONE')
df['Journal_title'] = df['Journal_title'].str.replace('Plos One', 'PLoS ONE')
df['Journal_title'] = df['Journal_title'].str.replace('PLoS One', 'PLoS ONE')


df.head()


```

```python
pl = df['Cost_(£)'].loc[df['Journal_title'] == 'PLoS ONE']
print('1) \"PLoS ONE\" with 178 Article title')
print('Mean of the open-access cost per article for \"PLoS ONE\" is: {}'.format(round(pl.mean(), 2)))
print('Median of the open-access cost per article for \"PLoS ONE\" is: {}'.format(round(pl.median(), 2)))
print('Standard deviation of the open-access cost per article for \"PLoS ONE\" is: {}'.format(round(pl.std(), 2)))
print('\n')
jbc = df['Cost_(£)'].loc[df['Journal_title'] == 'Journal of Biological Chemistry']
print('2) \"Journal of Biological Chemistry\" with 58 Article title')
print('Mean of the open-access cost per article for \"Journal of Biological Chemistry\" is: {}'.format(round(jbc.mean(), 2)))
print('Median of the open-access cost per article for \"Journal of Biological Chemistry\" is: {}'.format(round(jbc.median(), 2)))
print('Standard deviation of the open-access cost per article for \"Journal of Biological Chemistry\" is: {}'.format(round(jbc.std(), 2)))
print('\n')
pnas = df['Cost_(£)'].loc[df['Journal_title'] == 'Proceedings of the National Academy of Sciences']
print('3) \"Proceedings of the National Academy of Sciences\" with 28 Article title')
print('Mean of the open-access cost per article for \"PNAS\" is: {}'.format(round(pnas.mean(), 2)))
print('Median of the open-access cost per article for \"PNAS\" is: {}'.format(round(pnas.median(), 2)))
print('Standard deviation of the open-access cost per article for \"PNAS\" is: {}'.format(round(pnas.std(), 2)))
print('\n')
nar = df['Cost_(£)'].loc[df['Journal_title'] == 'Nucleic Acids Research']
print('4) \"Nucleic Acids Research\" with 24 Article title')
print('Mean of the open-access cost per article for \"Nucleic Acids Research\" is: {}'.format(round(nar.mean(), 2)))
print('Median of the open-access cost per article for \"Nucleic Acids Research\" is: {}'.format(round(nar.median(), 2)))
print('Standard deviation of the open-access cost per article for \"Nucleic Acids Research\" is: {}'.format(round(nar.std(), 2)))
print('\n')
hmg = df['Cost_(£)'].loc[df['Journal_title'] == 'Human Molecular Genetics']
print('5) \"Human Molecular Genetics\" with 178 Article title')
print('Mean of the open-access cost per article for \"Human Molecular Genetics\" is: {}'.format(round(hmg.mean(), 2)))
print('Median of the open-access cost per article for \"Human Molecular Genetics\" is: {}'.format(round(hmg.median(), 2)))
print('Standard deviation of the open-access cost per article for \"Human Molecular Genetics\" is: {}'.format(round(hmg.std(), 2)))

```
