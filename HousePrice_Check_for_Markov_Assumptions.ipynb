```python
# converting street to dummy variables
hprice = pd.concat([hprice,pd.get_dummies(hprice.street, drop_first=True)], axis=1)

dummy_columns = list(pd.get_dummies(hprice.street, drop_first=True).columns)
```

```python
hprice[['overallqual', 'overallcond', 'grlivarea'] + dummy_columns].head()
```

Now we use these variable in our model:

```python
X = hprice[['overallqual', 'overallcond', 'grlivarea'] + dummy_columns]
Y = hprice['saleprice']

lrm = linear_model.LinearRegression()

# fit method estimates the coefficients using OLS
lrm.fit(X, Y)

# Inspect the results.
print('\nCoefficients: \n', lrm.coef_)
print('\nIntercept: \n', lrm.intercept_)
```

```python
# Or with statsmodel
X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

Checking if the model meets the Gauss-Markov Conditions:
Y = - 0.0009 + 3.29X1 + 332X2 + 55.9X3 - 0.0001X4 

1) linearity of the model in its coefficients: meets

2) 𝔼(𝜖)=0: meets


```python
predictions = results.predict(X)
errors = Y - predictions
print("Mean of the errors in the weather model is: {}".format(np.mean(errors)))
```

It's almos equal to zero. Since, we include the constant term in the model, the average of the model's error is effectively zero.

3) homoscedasticity:

```python
plt.scatter(predictions, errors)
plt.xlabel('Predicted')
plt.ylabel('Residual')
plt.axhline(y=0)
plt.title('Residual vs. Predicted')
plt.show()
```

It seems that error terms almost are consistently distributed, and the model is homoscedastic.

4) Low molticollinearity:


```python
X.loc[:, X.columns != 'const'].corr()
```

Except for overallqual & grlivarea, there is no high correlation between variables

5) Error terms should be uncorrelated with one another:

```python
plt.plot(errors)
plt.show()
```

```python
from statsmodels.tsa.stattools import acf

acf_data = acf(errors)

plt.plot(acf_data[1:])
plt.show()
```

The autocorrelation between the errors of our medical costs model is indeed very low.

6) Features shouldn't be correlated with the errors:

```python
Z = hprice[['overallqual', 'overallcond', 'grlivarea', 'Pave']] 
Z['errors'] = errors
Z.corr()
```

The correalations between features and errors are very small, almost zero.

Check for normality:

```python
rand_nums = np.random.normal(np.mean(errors), np.std(errors), len(errors))

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(np.sort(rand_nums), np.sort(errors)) # we sort the arrays
plt.xlabel("the normally distributed random variable")
plt.ylabel("errors of the model")
plt.title("QQ plot")

plt.subplot(1,2,2)
plt.hist(errors)
plt.xlabel("errors")
plt.title("Histogram of the errors")

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import jarque_bera
from scipy.stats import normaltest

jb_stats = jarque_bera(errors)
norm_stats = normaltest(errors)

print("Jarque-Bera test statistics is {0} and p value is {1}".format(jb_stats[0], jb_stats[1]))
print("Normality test statistics is {0} and p value is {1}".format(norm_stats[0], norm_stats[1]))
```

Graphs and tests results indicate that our errors are not normally distributed.
